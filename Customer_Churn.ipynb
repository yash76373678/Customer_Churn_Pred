#### Now we see geography wise data and see total_customer and exited customer

```python
df1=pd.DataFrame()
df1['Total_Cust']=df.groupby('Geography').CustomerId.count()
df1['Exited']=df[df['Exited']==1].groupby('Geography').CustomerId.count()
df1['Churn_rate']=df1['Exited']/df1['Total_Cust']
df1
```

#### We can see that majority customer who left the bank are from Germany

#### Now we see gender wise distribution and how many are left from that 

```python
df2=pd.DataFrame()
df2['Gender']=df.groupby('Gender').CustomerId.count()
df2['Exited']=df[df['Exited']==1].groupby('Gender').CustomerId.count()
df2['Churn_rate']=df2['Exited']/df2['Gender']
df2
```

#### We can see that female are more exited from bank than male

#### Now we can use cut function to group age column

```python
df['Age_group']=pd.cut(x=df['Age'],bins=[18,30,40,50,60,70,80,90,100],labels=['18 to 30','30 to 40','40 to 50','50 to 60','60 to 70','70 to 80','80 to 90','90 to 100'])
```

```python
df.head()
```

#### Now we will see age_group wise how many customer exited from bank 

```python
df3=pd.DataFrame()
df3['Age_Group']=df.groupby('Age_group').CustomerId.count()
df3['Exited']=df[df['Exited']==1].groupby('Age_group').CustomerId.count()
df3['Churn_Rate']=df3['Exited']/df3['Age_Group']
df3.sort_values(by='Churn_Rate',ascending=False)
```

#### we can clearly see majority are from 50-60 age who left the bank we assume that bcz of mortality

#### Now we will see if tenure will affect the churn_rate or not

```python
df4=pd.DataFrame()
df4['Tenure']=df.groupby('Tenure').CustomerId.count()
df4['Exited']=df[df['Exited']==1].groupby('Tenure').CustomerId.count()
df4['Churn_rate']=df4['Exited']/df4['Tenure']
df4
```

#### We can see that all values are almost equal to each other so tenure will not affect churn_rate 

#### Now we use cut function for group the balance column  

```python
df['Balance_Group']=pd.cut(x=df['Balance'],bins=[-1,50000,100000,150000,200000,260000],labels=['0 to 50000','50000 to 100000','100000 to 150000','150000 to 200000','200000 to 260000'])
```

```python
df.head()
```

#### We will see how many customer exited from diff balance_group

#### We can see that high earning people are more likily to leave the bank

#### Now we will analyze numofproducts column

```python
df6=pd.DataFrame()
df6['NumOfProducts']=df.groupby('NumOfProducts').CustomerId.count()
df6['Exited']=df[df['Exited']==1].groupby('NumOfProducts').CustomerId.count()
df6['Churn_rate']=df6['Exited']/df6['NumOfProducts']
df6
```

```python
df7=pd.DataFrame()
df7['HasCrCard']=df.groupby('HasCrCard').CustomerId.count()
df7['Exited']=df[df['Exited']==1].groupby('HasCrCard').CustomerId.count()
df7['Churn_rate']=df7['Exited']/df7['HasCrCard']
df7
```

```python
df2=df[['Balance_Group','Geography','Complain','Gender','Age_group','NumOfProducts','HasCr','IsActiveMember','Exited']]
```

```python
X=df2.drop('Exited',axis=1)
y=df2['Exited']
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
categorical_columns = ['Balance_Group', 'Geography', 'Gender', 'Age_group']
X_encoded = pd.get_dummies(X, columns=categorical_columns)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=0)
```

```python
X_encoded
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
y_pred = dtree.predict(X_test)
print("Accuracy Score :", accuracy_score(y_test, y_pred)*100, "%")
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print("Accuracy Score :", accuracy_score(y_test, y_pred)*100, "%")
```
